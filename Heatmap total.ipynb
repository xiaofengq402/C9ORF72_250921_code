````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
import matplotlib.font_manager as fm

# 读取CSV文件
data = pd.read_csv('jjj.csv')

# 将基因名称设置为索引
data.set_index('Gene', inplace=True)

# 使用z-score进行归一化
normalized_data = data.apply(zscore, axis=1)

# 自定义颜色范围
desired_vmax = -0.5  # 最大值
desired_vmin = 2  # 最小值

if desired_vmax is None:
    desired_vmax = normalized_data.max().max()

# 设置字体大小（可以根据需要调整这里的值）
font_size = 7  # 示例字体大小设置

# 查找字体路径
font_path = fm.findfont(fm.FontProperties(family='Arial'))
font_prop = fm.FontProperties(fname=font_path, size=font_size)

# 绘制热图
plt.figure(figsize=(2.25, 2), dpi=200)  # 设置图像大小

# 定义颜色条的参数
cbar_kws = {
    'label': '',
    'shrink': 0.5,  # 缩小颜色条，取值范围 0-1，这里设置为 0.5 表示缩小到原来的一半
    'ticks': []  # 不显示颜色条的刻度
}

ax = sns.heatmap(
    normalized_data,
    annot=False, 
    cmap='Reds',  
    cbar_kws=cbar_kws, 
    vmin=desired_vmin,  
    vmax=desired_vmax,   
    linecolor='white',   
    linewidth=0.5        
) 

# 不显示y轴刻度和标签
ax.set_yticks([])  # 不显示y轴刻度
ax.set_yticklabels([])  # 不显示y轴标签（基因名称）

# 隐藏y轴的标签
ax.set_ylabel('')  # 清空y轴标签

# 不显示x轴刻度和标签
ax.set_xlabel('', fontsize=font_size)  # 清空x轴标签
ax.set_xticks([])  # 不显示x轴刻度
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: ''))  # 隐藏x轴的数字刻度

plt.title('', fontsize=font_size)  # 设置标题

plt.tight_layout()  # 自动调整布局以避免重叠
# 保存为PDF和TIFF格式
plt.savefig('heatmap.pdf', format='pdf', dpi=600)
plt.savefig('heatmap.tiff', format='tiff', dpi=600)

plt.show()

````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
import matplotlib.font_manager as fm
from matplotlib.colors import LinearSegmentedColormap
# 读取CSV文件
data = pd.read_csv('base_ISGs.csv')

# 将基因名称设置为索引
data.set_index('Gene', inplace=True)

# 使用z-score进行归一化
normalized_data = data.apply(zscore, axis=1)

# 自定义颜色范围
desired_vmax = -1.5  # 最大值
desired_vmin = 1.5  # 最小值

if desired_vmax is None:
    desired_vmax = normalized_data.max().max()

# 设置字体大小（可以根据需要调整这里的值）
font_size = 7  # 示例字体大小设置

# 查找字体路径
font_path = fm.findfont(fm.FontProperties(family='Arial'))
font_prop = fm.FontProperties(fname=font_path, size=font_size)

# 绘制热图
plt.figure(figsize=(2.5, 3.25),dpi=150)  # 设置图像大小

# 定义颜色条的参数
cbar_kws = {
    'label': '',
    'shrink': 0.5,  # 缩小颜色条，取值范围 0-1，这里设置为 0.5 表示缩小到原来的一半
    'ticks': []  # 不显示颜色条的刻度
}

# 创建红白蓝渐变色带
colors = ['#0606FF', 'white', '#FF0909']  # 天蓝 -> 白 -> 珊瑚红
cmap = LinearSegmentedColormap.from_list('rwb', colors, N=256)

# 修改heatmap参数部分
ax = sns.heatmap(
    normalized_data,
    annot=False,
    cmap=cmap,          # 替换颜色映射
    cbar_kws=cbar_kws,
    vmin=desired_vmin,  # 修正后的最小值
    vmax=desired_vmax,  # 修正后的最大值
    linecolor='white',  # 加深网格线颜色
    linewidth=0.5       # 适当减小线宽
)

# 不显示y轴刻度和标签
ax.set_yticks([])  # 不显示y轴刻度
ax.set_yticklabels([])  # 不显示y轴标签（基因名称）

# 隐藏y轴的标签
ax.set_ylabel('')  # 清空y轴标签

# 不显示x轴刻度和标签
ax.set_xlabel('', fontsize=font_size)  # 清空x轴标签
ax.set_xticks([])  # 不显示x轴刻度
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: ''))  # 隐藏x轴的数字刻度

plt.title('', fontsize=font_size)  # 设置标题

plt.tight_layout()  # 自动调整布局以避免重叠
# 保存为PDF和TIFF格式
plt.savefig('heatmap.pdf', format='pdf', dpi=600)
plt.savefig('heatmap.tiff', format='tiff', dpi=600)

plt.show()

````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np

# 读取 CSV 文件，不跳过空行
with open('jzl0905.csv', 'r') as f:
    lines = f.readlines()

# 分割数据块
blocks = []
current_block = []
for line in lines:
    if line.strip() == '' or line.startswith(',' * 9):  # 空行或只有逗号的行
        if current_block:
            blocks.append(current_block)
            current_block = []
    else:
        current_block.append(line.strip())

# 添加最后一个块
if current_block:
    blocks.append(current_block)

# 自定义颜色范围
desired_vmax = 5  # 最大值，根据需要调整
desired_vmin = 1  # 最小值，根据需要调整

# 设置字体大小
font_size = 7  # 示例字体大小设置

# 查找字体路径
try:
    font_path = fm.findfont(fm.FontProperties(family='Arial'))
    font_prop = fm.FontProperties(fname=font_path, size=font_size)
    # 创建斜体字体属性
    italic_font_prop = fm.FontProperties(fname=font_path, size=font_size, style='italic')
except:
    # 如果找不到 Arial 字体，使用默认字体
    font_prop = fm.FontProperties(size=font_size)
    italic_font_prop = fm.FontProperties(size=font_size, style='italic')

# 循环处理每个数据块
for i, block in enumerate(blocks):
    # 将块转换为数据框
    from io import StringIO
    block_str = '\n'.join(block)
    sub_table_data = pd.read_csv(StringIO(block_str))
    
    # 检查是否包含 'gene' 列
    if 'gene' not in sub_table_data.columns:
        print(f"子表 {i+1} 中未找到 'gene' 列，跳过处理")
        continue
    
    # 将 gene 列设置为索引
    sub_table_data.set_index('gene', inplace=True)
    
    # 确保索引是字符串类型，避免NaN问题
    sub_table_data.index = sub_table_data.index.astype(str)
    
    # 移除可能存在的空行
    sub_table_data = sub_table_data[sub_table_data.index != 'nan']
    
    # 将数据转换为数值类型
    sub_table_data = sub_table_data.apply(pd.to_numeric, errors='coerce')
    
    # 检查数据是否有效
    if sub_table_data.empty:
        continue
    
    # 创建图形和轴对象
    fig, ax = plt.subplots(figsize=(3, 1.88), dpi=200)
    
    # 定义颜色条的参数
    cbar_kws = {
        'label': '',
        'shrink': 0.5,
        'ticks': []
    }
    
    # 创建mask来隐藏NaN值
    mask = np.isnan(sub_table_data.values)
    
    # 绘制热图
    sns.heatmap(
        sub_table_data,
        annot=False,
        cmap='Blues',
        cbar_kws=cbar_kws,
        vmin=desired_vmin,
        vmax=desired_vmax,
        linecolor='white',
        linewidth=0.5,
        mask=mask,
        ax=ax
    )
    
    # 显示 y 轴刻度和标签（基因名）- 使用斜体
    ax.set_yticks(np.arange(len(sub_table_data.index)) + 0.5)  # 居中对齐
    ax.set_yticklabels(sub_table_data.index, fontsize=font_size, fontproperties=italic_font_prop)
    
    # 显示 x 轴刻度和标签（列名，即组别）- 居中对齐
    ax.set_xticks(np.arange(len(sub_table_data.columns)) + 0.5)  # 居中对齐
    ax.set_xticklabels(sub_table_data.columns, fontsize=font_size, rotation=45, 
                       ha='right', fontproperties=font_prop)
    
    # 调整x轴标签位置，避免与刻度线重叠
    ax.tick_params(axis='x', which='major', pad=8)  # 增加标签与刻度线的距离
    
    # 设置轴标签
    ax.set_ylabel('Gene', fontsize=font_size, fontproperties=font_prop, labelpad=10)
    ax.set_xlabel('Group', fontsize=font_size, fontproperties=font_prop, labelpad=10)
    
    # 设置标题
    ax.set_title(f'Heatmap {i + 1}', fontsize=font_size, fontproperties=font_prop, pad=10)
    
    # 调整布局，确保所有元素都可见
    plt.tight_layout()
    
    # 进一步调整布局，为x轴标签留出更多空间
    plt.subplots_adjust(bottom=0.1)
    
    # 保存为 PDF 和 TIFF 格式
    plt.savefig(f'heatmap_{i + 1}.pdf', format='pdf', dpi=600, bbox_inches='tight')
    plt.savefig(f'heatmap_{i + 1}.tiff', format='tiff', dpi=600, bbox_inches='tight')
    
    plt.show()
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
import matplotlib.font_manager as fm
import numpy as np

# 读取Excel文件
file_path = 'QXFstsum datum20252.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')

# 将基因名称设置为索引
data.set_index('Gene', inplace=True)

# 使用z-score进行归一化
normalized_data = data.apply(zscore, axis=1)

# 自定义颜色范围
desired_vmax = 2.5  # 最大值
desired_vmin = -2.5 # 最小值

if desired_vmax is None:
    desired_vmax = normalized_data.max().max()

# 设置字体大小
font_size = 7

# 查找字体路径
font_path = fm.findfont(fm.FontProperties(family='Arial'))
font_prop = fm.FontProperties(fname=font_path, size=font_size)

# 绘制热图 - 调整图形大小以适应标签
plt.figure(figsize=(4, 4), dpi=200)  # 增加图形尺寸以容纳标签

# 定义颜色条的参数
cbar_kws = {
    'label': 'Z-score',
    'shrink': 0.5,
    'ticks': np.linspace(desired_vmin, desired_vmax, 5)  # 添加颜色条刻度
}

ax = sns.heatmap(
    normalized_data,
    annot=False, 
    cmap='RdBu_r',  
    cbar_kws=cbar_kws, 
    vmin=desired_vmin,  
    vmax=desired_vmax,   
    linecolor='white',   
    linewidth=0.5        
) 

# 显示y轴基因名称
plt.yticks(rotation=0, fontsize=font_size)

# 显示x轴组别名称
plt.xticks(rotation=45, ha='right', fontsize=font_size)

# 设置轴标签
plt.xlabel('Sample Groups', fontsize=font_size+2)
plt.ylabel('Genes', fontsize=font_size+2)

# 调整布局以避免标签被截断
plt.tight_layout()

# 保存为PDF和TIFF格式
plt.savefig('heatmap.pdf', format='pdf', dpi=600, bbox_inches='tight')
plt.savefig('heatmap.tiff', format='tiff', dpi=600, bbox_inches='tight')

plt.show()
````
